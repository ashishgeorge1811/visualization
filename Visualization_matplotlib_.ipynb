# Annotation  for object oriented

```python
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([0,0,2,2]) # axes is internal length
#ax = fig.add_axes([.1,.1,1,1]) # axes is internal length
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names,rotation=120,fontsize = 15)# rotation is the angle
ax.set_title('Miles per gallon of cars in mtcars',fontsize = 25)
ax.set_xlabel('car names',fontsize = 15)
ax.set_ylabel('miles/gal',fontsize = 25)
ax.plot(mpg)
ax.annotate('Toyota corolla',xy = (19,33.9),xytext = (23,30),arrowprops=dict(facecolor= 'green'))
ax.annotate('Merc 230',xy = (7,25),xytext = (9,27),arrowprops=dict(facecolor= 'blue'))
```
